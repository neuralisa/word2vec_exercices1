```python
# Importing necessary libraries
import gensim.downloader as api
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import numpy as np

# Download the pretrained Word2Vec model (GoogleNews dataset)
model = api.load('word2vec-google-news-300')

# Function to plot words in 2D space
def plot_words_2d(model, words):
    word_vectors = np.array([model[word] for word in words])
    
    # Reduce dimensionality with PCA
    pca = PCA(n_components=2)
    word_vectors_pca = pca.fit_transform(word_vectors)
    
    # Plotting
    plt.figure(figsize=(10, 8))
    plt.scatter(word_vectors_pca[:, 0], word_vectors_pca[:, 1], c='red')
    
    for i, word in enumerate(words):
        plt.annotate(word, (word_vectors_pca[i, 0], word_vectors_pca[i, 1]))
    
    plt.title('Word2Vec Embeddings in 2D Space (PCA)')
    plt.show()

# Select some words to visualize
words = ['king', 'queen', 'man', 'woman', 'apple', 'orange', 'cat', 'dog', 'car', 'train']

# Plot the words using the model
plot_words_2d(model, words)
```
